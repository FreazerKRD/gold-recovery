### Проверим корректность расчета эффективности обогащения.

```python
# Создадим функцию для расчёта эфективности обогащения
def recovery(C, F, T):
    result = (C * (F - T)) / (F * (C - T)) * 100
    return result

# Используем функцию и сохраним результат в новом столбце
train['recovery'] = recovery(train['rougher.output.concentrate_au'], 
                             train['rougher.input.feed_au'], 
                             train['rougher.output.tail_au']
                            )
```

```python
# Посмотрим на проверяемый и проверочный столбцы
train[['rougher.output.recovery','recovery']]
```

```python
# Выведем сумму обоих столбцов
train['rougher.output.recovery'].sum(), train['recovery'].sum()
```

```python
# Посмотрим среднее абсолютное отклонение
mean_absolute_error(train['rougher.output.recovery'], train['recovery'])
```

```python
# Удалим созданный столбец
train.drop('recovery', axis=1, inplace=True)
```

**Эффективность обогащения в данных рассчитана верно, столбцы идентичны. MAE практически равно нулю.**

#### Сравним распределения размеров гранул сырья на обучающей и тестовой выборках до обогащения.

```python
# Построим гистограмму
plt.figure(figsize=(15,7))
sns.kdeplot(train['rougher.input.feed_size'], shade=True)
sns.kdeplot(test['rougher.input.feed_size'], shade=True)
plt.title('Размеры гранул сырья')
plt.xlabel('Значение признака')
plt.ylabel('Количество')
plt.legend(['Тренировочная выборка', 'Тестовая выборка'])
plt.show()

# Построим boxplot
plt.figure(figsize=(15,7))
sns.boxplot(data=[train['rougher.input.feed_size'], test['rougher.input.feed_size']],orient='h')
plt.grid()
plt.xlabel('Значение признака')
plt.title('Размеры гранул сырья')
plt.legend(['0 - Тренировочная выборка', '1 - Тестовая выборка'])
plt.show()

# Сделаем тест Колмогорова-Смирнова для проверки гипотезы о том, что выборки сделаны из одного распределения
stats.ks_2samp(train['rougher.input.feed_size'], test['rougher.input.feed_size'])
```

### Исследуем суммарную концентрацию всех веществ на разных стадиях: в сырье, в черновом и финальном концентратах.

```python
# Сохраним суммы всех концентратов
input_sum = full_df[['rougher.input.feed_ag', 'rougher.input.feed_pb', 'rougher.input.feed_sol', 'rougher.input.feed_au']].sum(axis=1)
draft_sum = full_df[['rougher.output.concentrate_ag', 'rougher.output.concentrate_pb', 'rougher.output.concentrate_sol', 'rougher.output.concentrate_au']].sum(axis=1)
final_sum = full_df[['final.output.concentrate_ag', 'final.output.concentrate_pb', 'final.output.concentrate_sol', 'final.output.concentrate_au']].sum(axis=1)

# Построим график
plt.figure(figsize=(15,9))
input_sum.hist(bins=101, color='red', alpha=0.6)
draft_sum.hist(bins=101, color='yellow', alpha=0.6)
final_sum.hist(bins=101, color='green', alpha=0.6)
plt.title('Суммарная концентрация металлов на разных стадиях')
plt.xlabel('Суммарная концентрация')
plt.ylabel('Количество')
plt.legend(['Сырьё', 'Черновой', 'Финальный'])
plt.show()

# Построим boxplot
plt.figure(figsize=(15,7))
sns.boxplot(data=[input_sum, draft_sum, final_sum],orient='h')
plt.grid()
plt.xlabel('Значение признака')
plt.title('Суммарная концентрация')
plt.legend(['0 - Сырьё', '1 - Черновой', '2 - Финальный'])
plt.show()
```

**График показывает, что в сырье концентрация имеет наибольший разброс значений, после обогащения суммарная концентрация металлов снижается почти в два раза, однако разброс значений становится заметно меньшим. После двухэтапной очистки суммарная концентрация сильно увеличивается, становясь выше, чем на всех остальных этапах и разброс значений становится еще ниже.**

**Построенные графики свидетельствуют о наличии некоторого количества объектов с околонулевой суммарной концентрацией. Скорее всего, это говорит нам о сбоях в работе оборудования или хранения данных.** 

**Устраним объекты, у которых суммарная концентрация веществ в сырье меньше края левого уса графика. Удалять данные по суммарной концентрации в черновом и финальном концентрате не будем, так как эти признаки не используются при обучении и использовании моделей.**

```python
# Расчитаем значение края левого уса
def del_outliers(sum, cols):
    # Расчитаем значение края левого уса
    median = np.median(sum)
    upper_quartile = np.percentile(sum, 75)
    lower_quartile = np.percentile(sum, 25)

    iqr = upper_quartile - lower_quartile
    lower_whisker = lower_quartile - 1.5 * iqr
    
    # Удалим значения из всех датафреймов
    global full_df, train, test
    full_df = full_df.loc[full_df[cols].sum(axis=1) > lower_whisker]
    train = train.loc[train[cols].sum(axis=1) > lower_whisker]
    try:
        test = test.loc[test[cols].sum(axis=1) > lower_whisker]
    except:
        print('Columns not in "test"')
```

### Создадим функцию для расчета метрики sMAPE.

```python
# Сделаем функцию для расчета метрики
def smape(target, predictions):
    result = abs(target - predictions) / ((abs(target) + abs(predictions)) / 2) * 100
    result = result.fillna(value=0)

    result = np.mean(result)
    return result

smape_scorer = make_scorer(smape, greater_is_better=False)

# Сделаем функцию для расчета итоговой метрики
def total_smape(smape_rougher, smape_final):
    return (0.25 * smape_rougher + 0.75 * smape_final)
```

```python
# Обучим модель
cb_rougher = cb.CatBoostRegressor(loss_function='MAPE', eval_metric='SMAPE')
cb_rougher.fit(train_features, train_rougher)

# Получим метрику smape для предсказаний 'rougher.output.recovery'
rougher_predictions = cb_rougher.predict(test_features)
print('Метрика SMAPE для предсказаний: ', smape(test_rougher, rougher_predictions))

# Сохраним предсказания в фичи для второй стадии проверки модели
test_features['cb_rougher.predict'] = rougher_predictions
train_features['cb_rougher.predict'] = cb_rougher.predict(train_features)

# Обучим дамми-модель
dummy_model = DummyRegressor(strategy="mean")
dummy_model.fit(train_features, train_rougher)

# Получим метрику smape для dummy-предсказаний 'rougher.output.recovery'
dummy_rougher_predictions = dummy_model.predict(test_features)
print('Метрика SMAPE для dummy: ', smape(test_rougher, dummy_rougher_predictions))
```

```python
# Обучим модель
cb_final = cb.CatBoostRegressor(loss_function='MAPE', eval_metric='SMAPE')
cb_final.fit(train_features, train_final)

# Получим предсказания второго целевого признака 'final.output.recovery'
final_predictions = cb_final.predict(test_features)
print('Метрика SMAPE для предсказаний: ', smape(test_final, final_predictions))

# Обучим дамми-модель
dummy_model.fit(train_features, train_final)

# Получим метрику smape для dummy-предсказаний 'rougher.output.recovery'
dummy_final_predictions = dummy_model.predict(test_features)
print('Метрика SMAPE для dummy: ', smape(test_final, dummy_final_predictions))
```

```python
# Посчитаем итоговое sMAPE для модели CatBoostRegressor
print('Итоговое sMAPE CatBoost:', total_smape(smape(test_rougher, rougher_predictions), smape(test_final, final_predictions)))

# Посчитаем итоговое sMAPE для модели dummy
print('Итоговое sMAPE Dummy:', total_smape(smape(test_rougher, dummy_rougher_predictions), smape(test_final, dummy_final_predictions)))
```

**Выводы:** проверяемая модель показала результат метрики sMAPE в 5.36. При этом оценка sMAPE константной модели составила 6.52, что говорит о вменяемости нашей модели.
